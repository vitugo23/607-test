```python
# Load Libraries required for the project
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Create a Graph function
G = nx.Graph()
```

```python
# Add nodes to graph based on sample
G.add_nodes_from(["Andre", "Carol", "Fernando", "Garth", "Beverly", "Ed", "Heather", "Ike", "Jane"])
```

```python
# Add edges based on sample
G.add_edges_from([("Andre", "Carol"), ("Andre", "Fernando"), ("Carol", "Fernando"), ("Andre", "Beverly"),
                  ("Fernando","Garth"),("Fernando","Heather"),
                  ("Garth","Heather"),
                  ("Beverly","Garth"),("Beverly","Ed"),
                  ("Ed","Garth"),
                  ("Heather","Ike"),
                  ("Ike","Jane"),
                  ("Diane","Andre"),("Diane","Carol"),("Diane","Fernando"),("Diane","Garth"),("Diane","Ed"),("Diane","Beverly")])
```

```python
# Create a layout
pos = {"Andre": (0, 6), "Carol": (1, 8), "Fernando": (2,6), "Garth": (2, 2), "Beverly" : (0,2),
       "Diane" : (1,4),
      "Ed" : (1,0),
      "Heather" : (4,4), "Ike" : (6,4), "Jane" : (8,4)} 
```

```python
# Design the drawing
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=12, font_color='black')
```

```python
# Display the graph
plt.show()
```
